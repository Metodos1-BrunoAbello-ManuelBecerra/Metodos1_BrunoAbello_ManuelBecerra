```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from tqdm import tqdm
```

```python
def f(x):
    return np.abs(x-2)
```

```python
Dx = lambda f,x,h=1e-5: (f(x+h) - f(x-h))/(2*h)
```

```python
def Minimizer(f, N=400, gamma=0.01):
    
    r = np.zeros(N)
    # Seed
    r[0] = np.random.uniform(-5,5)
    
    for i in tqdm(range(1,N)):
        r[i] = r[i-1] - gamma*Dx(f,r[i-1])
        
    return r
```

```python
x = Minimizer(f)
```

```python
_x = np.linspace(0,3,100)
```

```python
plt.plot(_x, f(_x))
plt.axvline(x=x[-1])
plt.show()
#plt.clf()
```
